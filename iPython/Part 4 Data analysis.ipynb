```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# High heat map
def height_heatmap(height_path):
    data = pd.read_csv(height_path,index_col=0)
    np.set_printoptions(precision=2)
    data=np.array(data['10%'])
    dt = pd.DataFrame(data=(data[0:15],data[15:30],data[30:45],data[45:60],data[60:75],data[75:90],
                       data[90:105],data[105:120],data[120:135],data[135:150],data[150:165],data[165:180],
                       data[180:195],data[195:210],data[210:225],data[225:240],data[240:255],data[255:270]))
    fig, ax = plt.subplots(figsize=(5,10))
    ax = sns.heatmap(dt,linewidths=.5, vmin=0, vmax=145,cmap='jet',xticklabels=False, yticklabels=False,fmt ='.3',
                    cbar_kws={"shrink": 0.5})
```
